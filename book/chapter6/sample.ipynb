```python
df = pd.DataFrame(
[
(1, "202004", 10, 'indihome'),
(1, "202004", 20, 'indihome'),
(2, "202004", 40, 'myrepublic'),
(2, "202005", 5, 'myrepublic'),
(2, "202005", 100, 'xlhome'),
(2, "202005", 20, 'xlhome'),
(2, "202005", 40, 'xlhome'),
(2, "202005", 80 , 'myrepublicindonesia'),
(4, "202005", 90, 'myrepublicindonesia'),
(4, "202005", 10, 'abc'),
(5, "202005", 80, 'abc'),
(5, "202004", 50, 'abc'),
(4, "202004", 10, 'indihome')
]
,
columns = ["msisdn", "month", "trx", "apps"]

)
```

```python
required_list = ['indihome','myrepublic','xlhome','myrepublicindonesia','firstmedia']
```

```python
def cnt_func(d):
    cnt = 0
    for item in  d:
        if item in required_list:
            cnt = cnt + 1
    return cnt
    
    
```

```python
df_ = df.groupby(["msisdn", "month"])['apps'].agg(cnt_func)
```
